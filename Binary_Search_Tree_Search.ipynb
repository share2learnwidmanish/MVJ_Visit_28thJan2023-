## Create a class object to represent Node.
- Node will have 3 properties Key/Data, left reference, right reference 

```python
class BST: 
    def __init__(self,key):
        self.key = key 
        self.lchild = None
        self.rchild = None       
```

```python
root = BST(10)
print(root.key)
print(root.lchild)
print(root.rchild)
```

```python
root.lchild = BST(5)
print(root.lchild.key)
print(root.lchild.lchild)
print(root.lchild.lchild)
```

## Handling insert operation for empty tree
### To call object of a class you need an object of the class. For empty tree there will be no objects created. To handle this we areate a node object with "none" as the key 
root = BST(None) 
root.insert_node(10)
### and ignore any node with key as "None"
If a Tree contain only one node with key == "None" we can consider it as empty tree and can use this object(Node with None key) to call insert object

```python
class smart_BST: 
    def __init__(self,key):
        self.key = key 
        self.lchild = None
        self.rchild = None  
    
    def insert_node(self, data):
        ''' We need data for th enew node 
            First we need to check if the Tree is empty,
            Condition #1 : if key for first nNode is None we can assume tree is empty '''
        if self.key is None:
            # No need to create a new node, update the Data/key
            self.key = data
            return
        # If tree is not empty identify if it will go to left subtree or right subtree
        # If data of new node is < root node key => Left subtree
           ## Scenario #1 : Left Subtree can be empty, create a new node and add proper references
           ## Scenario #2 : Left Subtree is not empty, call insert method on root.lchild.insert(data)
        if self.key > data:
            if self.lchild:
                self.lchild.insert_node(data) # recursion
            else:
                self.lchild = smart_BST(data) # Will create a new node
        # Insert node on Right side 
        #If data of new node is > root node key => Right subtree
        else: 
            if self.rchild:
                self.rchild.insert_node(data) # recursion
            else:
                self.rchild = smart_BST(data) # Will create a new node
        # To handle duplicate values
        if self.key == data: 
            return
        
    def search(self, data):
            # Compare data with root node
            # Match => You found the data 
            # Data < root Key => Left subtree 
            # Data > root key => Right subtree
            if self.key == data:
                print("Node Found !")
                return
            if data < self.key:
                # Search in Left subtree 
                # Condition1: Left subtree can be empty then node is not present  
                # Condition2: Left subtree can have 1 or more nodes, recursive search operation
                if self.lchild:
                    self.lchild.search(data)
                else:
                    print("Node NOT Found in tree !")
            else:
                if self.rchild:
                    self.rchild.search(data)
                else:
                    print("Node NOT found in tree !")
            
```

```python
root = smart_BST(10)
```
